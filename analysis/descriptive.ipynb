#### 3. Segmentasi Pelanggan

```python
import pandas as pd
from datetime import datetime

# Asumsi df adalah gabungan tabel Orders dan Order_Details
# Kita tentukan 'snapshot_date' (biasanya hari ini atau hari terakhir di dataset)
snapshot_date = descriptive['tanggal_pesan'].max() + pd.Timedelta(days=1)

# Agregasi data per pelanggan
rfm = descriptive.groupby('id_pelanggan').agg({
    'tanggal_pesan': lambda x: (snapshot_date - x.max()).days, # Recency
    'id_pesan': 'nunique',                                     # Frequency
    'revenue': 'sum'                                           # Monetary
})

# Rename kolom
rfm.rename(columns={
    'tanggal_pesan': 'Recency',
    'id_pesan': 'Frequency',
    'revenue': 'Monetary'
}, inplace=True)
```

```python
# Skor 5 adalah yang terbaik (Recency rendah = skor tinggi, F & M tinggi = skor tinggi)
rfm['R_score'] = pd.qcut(rfm['Recency'], 5, labels=[5, 4, 3, 2, 1])
rfm['F_score'] = pd.qcut(rfm['Frequency'].rank(method='first'), 5, labels=[1, 2, 3, 4, 5])
rfm['M_score'] = pd.qcut(rfm['Monetary'], 5, labels=[1, 2, 3, 4, 5])

# Gabungkan jadi satu skor RFM
rfm['RFM_Group'] = rfm.R_score.astype(str) + rfm.F_score.astype(str) + rfm.M_score.astype(str)
rfm
```

```python
# Mapping menggunakan regex
segs = {
    r'[1-5][1-2]5':'Big Spenders',
    r'555': 'Best Customers',
    r'[4-5][4-5][1-2]':'Cheap Loyalist',
    r'51[4-5]':'New Big Spenders',
    r'[1-2][3-5][4-5]':'High Value at Risk',
    r'333':'Average Joe',
    r'11[1-2]':'One Time Low Spender',
    r'[4-5]13':'Promising'
}

# gabungkan skor
rfm['Segment'] = rfm['R_score'].astype(str) + rfm['F_score'].astype(str) + rfm['M_score'].astype(str)
rfm['Segment'] = rfm['Segment'].replace(segs, regex=True)
```

```python
rfm.loc[rfm['Segment'] == 'High Value at Risk']
```

Pelanggan-pelanggan tersebut dulu sangat loyal dan belanja banyak, tapi sekarang sudah lama tidak muncul. Ini prioritas nomor 1 untuk diselamatkan.

Pelanggan ini belanjanya biasa saja, frekuensi sedanng. Disarankan untuk memberikan penawaran untuk meningkatkan mereka jadi *loyal customers*
